```python
# importing necessary libraries
import pandas as pd 
import numpy as np
import seaborn as sns
import re

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
# Adjusts the configuration of the plots
matplotlib.rcParams['figure.figsize'] = (12,8)

# read in data from .csv file
df = pd.read_csv(r'C:\Users\Esteffy\Downloads\movies.csv')
```

```python
# let's look at the data !
df.head()
```

```python
# checking if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, (pct_missing)))
```

```python
# checking data types of columns

df.dtypes
```

```python
# replace all NaN elements w/ 0s since budget has some
df = df.fillna(0)

# changing data types for columns budget, gross to int64
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# creating a new column w/ correct release year

for col in df['released'].head():
    df['yearCorrect'] = re.sub(r"\([^()]*\)", "", col)
    df['yearCorrect'] = df['yearCorrect'].astype(str).str[-5:]

df
```

```python
# order by gross revenue
pd.set_option('display.max_rows', 20)
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# display options
pd.set_option('display.max_rows', 20)
```

```python
# scatter plot with BUDGET VS GROSS

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget')
plt.show()
```

```python
# conclusion: there is a high correlation between budget and gross
```

```python
# now let's look at company correlation

df_numerized = df

for col_name in df_numerized.columns:
    # if the column has an object data type, we want to change to category type
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        # this gives random numerization
        df_numerized[col_name] = df_numerized[col_name].cat.codes
    
df_numerized
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
# organize to see the highest correlation 

# display options
pd.set_option('display.max_rows', None)
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

sorted_corr_pairs = corr_pairs.sort_values()

sorted_corr_pairs
```

```python
high_corr = sorted_corr_pairs[(sorted_corr_pairs) > 0.5]

high_corr
```
